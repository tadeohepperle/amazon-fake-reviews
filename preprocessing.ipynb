```python
import pandas as pd
import numpy as np
import textblob
```

```python
string_sanetize = lambda x:  "mama" if str(x) == "nan" else ""
```

```python
df = pd.read_csv("data/amazon_reviews.txt", sep="\t")
df = df.rename(columns={"PRODUCT_ID": "product_id", "PRODUCT_TITLE": "product_title","REVIEW_TITLE": "review_title","REVIEW_TEXT": "review_text", "VERIFIED_PURCHASE": "verified", "RATING": "rating" })
df['product_id'] = df['product_id'].apply(lambda x: str(x)).astype(str)
df['product_title'] = df['product_title'].apply(string_sanetize).astype(str)
df['rating'] = df['rating'].apply(lambda x: float(x))
df = df.drop(columns=["DOC_ID"])
df
```

```python
reviews_scraped = pd.read_csv("data/product_reviews_scraped.csv", sep=",")
reviews_scraped["verified_purchase"] = np.where(reviews_scraped["verified_purchase"] == True, 1.0, 0.0)
reviews_scraped = reviews_scraped.rename(columns={"id": "product_id", "title": "review_title","content": "review_text", "verified_purchase": "verified", "rating": "rating" })
reviews_scraped['review_text'] = reviews_scraped['review_text'].apply(string_sanetize).astype(str)
reviews_scraped['product_id'] = reviews_scraped['product_id'].apply(string_sanetize).astype(str)
reviews_scraped
```

```python
products_scraped = pd.read_csv("data/products_scraped.csv", sep=",")
products_scraped = products_scraped.rename(columns={"title": "product_title", "id": "product_id"})
products_scraped['product_id'] = products_scraped['product_id'].apply(string_sanetize).astype(str)
products_scraped['product_title'] = products_scraped['product_title'].apply(string_sanetize).astype(str)
products_scraped
```

## Data Preprocessing

```python
# one hot encode product category:
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore')
encoder_df = pd.DataFrame(encoder.fit_transform(df[['PRODUCT_CATEGORY']]).toarray())
encoder_df = encoder_df.rename(lambda x: f"cat_{x}", axis=1)
df = df.join(encoder_df)
CATEGORY_COLUMNS = [e for e in encoder_df.columns]
# encode verified purchase as y=1, n=0
df["verified"] = np.where(df["verified"] == "Y", 1.0, 0.0)
df["label"] = np.where(df["LABEL"] == "__label1__", 1.0, 0.0)
```

```python
# export product ids:
f = open("data/ids.txt", "w+")
f.write("\n".join([e for e in df["product_id"]]))
f.close()
```

## Feature Engineering

```python
# sentiment analysis scores
from textblob import TextBlob

def get_polarity(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity

def get_subjectivity(text):
    blob = TextBlob(text)
    return blob.sentiment.subjectivity
```

```python
# Sentiment Score for Review tex in df
df['text_sentiment'] = df['review_text'].apply(get_polarity) +1 /2
df['text_subjectivity'] = df['review_text'].apply(get_subjectivity)
```

```python
# Sentiment Score for Review tex in df
reviews_scraped['text_sentiment'] = reviews_scraped['review_text'].apply(get_polarity)
reviews_scraped['text_subjectivity'] = reviews_scraped['review_text'].apply(get_subjectivity)
```

```python
df = df.merge(products_scraped, on=["product_id"])
```

```python
df["product_title"] = df["product_title_x"]
df = df.drop(columns=["LABEL", "product_title_x", "product_title_y"])
reviews_scraped = reviews_scraped.drop(columns=["rating_count", "rating_avg", "rating1", "rating2", "rating3", "rating4", "rating5"])
```

# Saving the data:

```python
df.to_csv("data/transformed/df.csv")
reviews_scraped.to_csv("data/transformed/reviews_scraped.csv")
```
